BASE DE DATOS

Becas nuevas movilidad al extranjero 2018

La presente base de datos nos da informacion del Listado de Becas Movilidad al Extranjero Nuevas por Área, Nivel,Institución, Entidad Federativa y Sexo 2018

0. Carga de librerias y base de datos

```python
import pandas as pd
b=pd.read_excel('/content/drive/MyDrive/Becas-Nuevas-Movilidad-Extranjero-2018.xlsx')
b
```

1. CONOCER LA CANTIDAD DE BECAS QUE FUERON OTORGADAS EN ESE AÑO

```python
b.NOMBRE.count()
```

2.CONOCER LAS AREAS DE LOS ASPIRANTES

```python
from pandas.core.groupby import groupby
A=b.groupby(['AREA'])['NOMBRE'].count()
A
```

EN LA PRIMER GRAFICA PODEMOS APRECIAR COMO SE DISTRIBUYEN LAS BECAS POR AREAS.
DONDE SE OBSERVA QUE LA MAYOR CANTIDAD DE BECAS SON PARA EL AREA DE SOCIALES, MIENTRAS QUE LA MENOR CANTIDAD ES DE FISICO MATEMATICAS


```python
import plotly.express as px
fig1 = px.bar(b, x = 'AREA', color = 'AREA', title='AREAS DE LOS BECADOS')
fig1
```

AHORA, APRECIAMOS LA DISTRIBUCION DE BECAS POR AREAS Y POR NIVELES

```python
import plotly.express as px
fig = px.bar(b, x="AREA", color="NIVEL", title='BECAS POR AREA Y NIVELES')
fig.show()
```

```python
fig = px.scatter_matrix(b, dimensions=["NIVEL", "PAÍS DESTINO"], color="AREA")
fig.show()
```

3. CONOCER EL NIVEL EN QUE FUE OTORGAGADA LA BECA

```python
b.groupby(['NIVEL'])['NOMBRE'].count()
```

EN LA SIGUIENTE GRAFICA PODEMOS APRECIAR QUE LA MAYOR CANTIDAD DE BECAS SON PARA ESTUDIANTES DE MAESTRIA, LE SIGUE DOCTORADO Y ESPECIALIDAD, MIENTRAS QUE LA MENOR ES PARA ESTANCIAS TECNICAS

```python
import plotly.express as px
fig = px.bar(b, x="NIVEL", title='Cantidad de becas por nivel')
fig.show()
```

```python
fig = px.pie(b,"NIVEL", title='Cantidad de becas por nivel')
fig.show()
```

4. CONOCER EL PAIS DE DESTINO DE LOS JOVENES BECADOS

```python
b.groupby(['PAÍS DESTINO'])['NOMBRE'].count()
```

EN EL SIGUIENTE HISTOGRAMA PODEMOS APRECIAR QUE LA MAYORIA DE LOS JOVENES BECADOS ESTUDIAN EN PAISES HISPANO HABLANTES COMO LO ES ESPAÑA, ARGENTINA, COLOMNIA, AUNQUE TAMBIEN SE APRECIAN BASTANTES PAISES EUROPEOS.
TAMBIEN PODEMOS OBSERVAR QUE LOS ESTUDIANTES DE INTERCAMBIO DE MAESTRIA Y DOCTORADO PREFIEREN ESPAÑA, LOS DE ESTANCIA TECNICA PREFIEREN JAPON.

```python
fig = px.histogram(b,"PAÍS DESTINO", color='NIVEL', title='DISTRIBUCIÓN DE BECAS POR PAIS DE DESTINO Y NIVEL EDUCATIVO')
fig.show()
```

LOS ESTUDIANTES DE CIENCIAS SOCIALES REALIZAN MOVILIDAD A ESPAÑA, ESTADOS UNIDOS Y COLOMBIA, FISICO MATEMATICAS E INGENIERIA A ESPAÑA, REINO UNIDO Y ESTADOS UNIDOS

```python
fig = px.histogram(b,"PAÍS DESTINO", color='AREA', title='PAÍS DE DESTINO POR AREAS')
fig.show()
```

6.CONOCER EL SEXO DE LOS JOVENES BECADOS

```python
b.groupby(['SEXO'])['NOMBRE'].count()
```

LA CANTIDAD DE BECAS PARA HOMBRES Y MUJERES VARIA UN POCO INCLINADA AL SEXO MASCULINO, AUNQUE LA OPORTUNIDAD DE ADQUIRIR UNA BECA ES IGUAL

```python
fig = px.pie(b,"SEXO", title='PORCENTAJE POR SEXO DE LOS BECADOS')
fig.show()
```
